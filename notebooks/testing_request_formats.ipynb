```python
import requests
ACTIVATION_KEY = "bf27a9c3e0f9431f"
```

```python
LALAL_API_URL = "https://www.lalal.ai/api"
FILENAME = "../audio_files/Furthest Thing.mp3"

headers = {
    "Content-Disposition": f"attachment; filename={FILENAME}",
    "Authorization": f"license {ACTIVATION_KEY}"
}

with open(FILENAME, "rb") as f:
    upload_response = requests.post(
        f"{LALAL_API_URL}/upload/",
        headers=headers,
        files={
            "file": (FILENAME, f, "audio/mpeg")
        }
    )
```

```python
print(upload_response.json())
```

```python
split_headers = headers.copy()
split_headers.pop("Content-Disposition")
```

```python
import json

params = [{
    "id": upload_response.json()["id"],
    "splitter": "phoenix",
    "stem": "piano"
}]

data = {
    "params": json.dumps(params)
}

response = requests.post(
    f"{LALAL_API_URL}/split/",
    headers=headers,
    data=data
)
```

```python
id_data = {
    "id": upload_response.json()["id"]
}

result_response = requests.post(
    f"{LALAL_API_URL}/check/",
    headers=split_headers,
    data=id_data
)
result_response.json()
```

```python
import ray
from ray import serve

# Connect to the running Ray cluster
ray.init(address="auto", ignore_reinit_error=True)

# Check if Serve is running
try:
    # Get all deployments
    deployments = serve.list_deployments()
    print("Ray Serve is running!")
    print(f"Active deployments: {list(deployments.keys())}")
    
    # Get detailed info about a specific deployment
    if "InstrumentDetectJobProducer" in deployments:
        info = serve.get_deployment("InstrumentDetectJobProducer")
        print(f"\nDeployment info: {info}")
except Exception as e:
    print(f"Ray Serve is not running or not accessible: {e}")
```
